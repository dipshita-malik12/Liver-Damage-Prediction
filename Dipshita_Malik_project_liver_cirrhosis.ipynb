```python
#by using corr(co relation) to get relation between columns to remove unwanted columns

col_df = df[['N_Days', 'Age',  'Bilirubin','Cholesterol', 'Albumin', 'Copper',
       'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets', 'Prothrombin','Stage']]
col_df.corr()
```

```python
sns.heatmap(col_df.corr(),annot=True,cmap="Spectral")
```

```python
#to treat outliers in code we use this code

for c in col_df.columns:
    percentile25 = df[c].quantile(0.25)
    percentile75 = df[c].quantile(0.75)
    iqr = percentile75 - percentile25
    upper_limit = percentile75 + (1.5*iqr)
    lower_limit = percentile25 - (1.5*iqr)
    df = df[df[c]<upper_limit]
    df = df[df[c]>lower_limit]
    print(iqr)
    print(upper_limit)
    print(lower_limit)
    plt.figure()
    sns.boxplot(y=c,data=col_df)
```

```python
X = df[['N_Days', 'Status',  'Age', 'Sex', 'Ascites', 'Hepatomegaly',
       'Spiders', 'Edema', 'Bilirubin', 'Cholesterol', 'Albumin', 'Copper',
       'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets', 'Prothrombin',
       'Stage']]
```

```python


from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[("encoder",OneHotEncoder(sparse_output=False),[1,2,4,5,6,7,8])],remainder = "passthrough")
x = np.array(ct.fit_transform(X))
pd.DataFrame(X)
```

```python
#importing train_test_split module from sklearn.module_selection

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
x_train,y_train
```

```python
#importing standardscaler module from sklearn.preprocessing

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
sklearn.metrics.confusion_matrix(y_test,y_pred)
```

```python
#calculating accuracy

score =(accuracy_score(y_test,y_pred))*100
print('Accuracy of the model is ',score)
```

# TESTING

```python
import joblib
```

```python
y_prediction=[]
for i in range(x_test.shape[0]):
    check_data=x_test[i,0:]
    joblib.dump(reg, "logistic_regression_model.pkl")
    new_data=check_data.reshape(1,-1)
    model=joblib.load("logistic_regression_model.pkl")
    prediction=model.predict(new_data)
    y_prediction.append(prediction)
    if prediction == 1:
        y_prediction.append(1)
    else:
        y_prediction.append(0)
```

```python
y_actual=[]
for i in range(y_test.shape[0]):
    y_actual.append(y_test[i])
```

```python
A=pd.Series(y_actual)
B=pd.Series(y_prediction)
```

```python
pd.concat([A,B],axis=1).rename(columns={0:"Actual_Values", 1:"Predicted Values"})
```
